```python
def replace_(x):
    try:
        x=x.replace(" ", "")
    except:
        return x
    return x
```

```python
import json
import numpy as np

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)

with open('../output/patient_data.json', 'w') as outfile:
    json.dump(new_json, outfile,  cls = NpEncoder)
```
