Data Cleaning rules:

Clean up string: startwith P - Pathogenic; start with LP or Likely P -> Likely Pathogenic; startwith U - Uncertain Significance; startswith B - Benign; startswith Likely B or LB -> Likely Benign

```python
def cleansing(text):
    if text.startswith("P") == True:
        return "Pathogenic"
    elif text.startswith("LP") == True or text.startswith("Likely P") == True:
        return "Likely Pathogenic"
    elif text.startswith("U") == True:
        return "Uncertain Significance"
    elif text.startswith("B") == True:
        return "Benign"
    elif text.startswith("LB") == True or text.startswith("Likely B") == True:
        return "Likely Benign"
    else:
        print("Error")
        print(text)
        return text
```

```python
final["Final Call"] = final["Final Call"].apply(lambda x : cleansing(x))

final["Mutation_ID"] ='chr'+ final["Chromosome"].astype(str) + ':' + final["Position"].astype(str) +"_" + final['Reference Allele']+'-->'+final['Sample Allele']
table2_1["curated"] = "No"
final["curated"] = "Yes"               

change = final[["Mutation_ID","Final Call", "curated"]]

change_dict = change.set_index('Mutation_ID').to_dict()['Final Call']
change_dict_2 = change.set_index('Mutation_ID').to_dict()["curated"]
change_dict
# table2_1 = table2_1.replace(change_dict)


for mutation_ID in change_dict.keys():
#     table2_1[table2_1.Mutation_ID == mutation_ID]["Classification"] = change_dict[mutation_ID]
    try:
        table2_1.loc[mutation_ID,"Classification"] = change_dict[mutation_ID]
    except:
        continue
    try:
        table2_1.loc[mutation_ID,"curated"] = change_dict_2[mutation_ID]
    except:
        continue
    
    
```
