```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../data/Genomics/Genomics_5.xlsx", sheet_name='All Out')
```

```python
list(data.columns)
```

```python
table = data[['Chromosome',
 'Position',
 'Reference Allele',
 'Sample Allele',
 'Variation Type',
 'Gene Region',
 'Gene Symbol',
 'Transcript ID',
 'Transcript Variant',
 'Protein Variant',
 'Variant Findings']]
```

```python
table["Transcript ID"] = table["Transcript ID"].astype(str).apply(lambda x : x.replace(" ", "").split(';'))
table["Transcript Variant"] = table["Transcript Variant"].astype(str).apply(lambda x : x.replace(" ", "").split(';'))
table["Protein Variant"] = table["Protein Variant"].astype(str).apply(lambda x : x.replace(" ", "").split(';'))
```

```python
table["Mutation_ID"] ='chr'+ table["Chromosome"].astype(str) + ':' + table["Position"].astype(str) +"_" + table['Reference Allele']+'-->'+table['Sample Allele']
```

```python
import json
from tqdm import tqdm
new_json = []
for i in tqdm(range(len(table["Position"]))):
    line = table.iloc[i]
    data=json.loads(line.to_json())
#     print(data)
    new_json.append(data)
```

```python
import json
with open('../output/mutation_information_5.json', 'w') as outfile:
    json.dump(new_json, outfile)
```
